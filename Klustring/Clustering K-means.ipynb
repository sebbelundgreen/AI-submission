```python
## Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
##Importing dataset
dataset = pd.read_excel('Absenteeism_at_work.xls')

## Number of children + abscent hours/year
X = dataset.iloc[:, [13, 20]].values
```

```python
## Checking values
X
```

```python
## plotting
plt.scatter(X[:, 0], X[:, 1], s = 20, c = 'blue')
plt.title('Employee abscent')
plt.xlabel('Children')
plt.ylabel('Hours abscent (In a year)')
plt.show()
```

```python
## Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0)
```

```python
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 10, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 10, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 10, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=40,c='black',label='Centroids')
plt.title('Clusters of Employees')
plt.xlabel('Children at home')
plt.ylabel('Hours away (in a year)')
plt.legend()
plt.show()
```
